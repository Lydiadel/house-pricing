## Dependencies

```python
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
from collections import Counter

from pathlib import Path

import sqlite3
```

## Create the database

```python
## Create the database NAME: mock_database.sqlite
Path('mock_database.sqlite').touch()
```

```python
## Connecto to the database
conn = sqlite3.connect('mock_database.sqlite')
c = conn.cursor()
```

```python
## Create table named mock_data in database named mock_database.sqlite
c.execute('''CREATE TABLE mock_data_table (Id int, LotArea int, firstFlrSF int, GrLivArea int, RoofStyle text, SalePrice int)''')
```

```python
# load the data into a Pandas DataFrame
mockdata_del_csv = pd.read_csv('mock_data_csv.csv')
```

```python
# write the data to a sqlite table
mockdata_del_csv.to_sql('mock_data_table', conn, if_exists='append', index = False)
```

```python
## Check the data is in there in table table_users2

c.execute('''SELECT * FROM mock_data_table''').fetchall()
```

## Close connection to the database

```python
conn.close()
```
