## Dependencies

```python
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
from collections import Counter

from pathlib import Path

import sqlite3
```

## Create the database

```python
## Create the database NAME: db_w2tables
Path('db2_w2tables.sqlite').touch()
```

```python
## Connecto to the database
conn = sqlite3.connect('db2_w2tables.sqlite')
c = conn.cursor()
```

```python
## Create table named table_users2 in database named db_w2tables
c.execute('''CREATE TABLE table_users2 (user_id2 int, username2 text)''')

## Create table named table_city2 in database named db_w2tables
c.execute('''CREATE TABLE table_city2 (name2 text, population2 int)''')
```

```python
# load the data into a Pandas DataFrame
users2_del_csv = pd.read_csv('users2.csv')

# load the data into a Pandas DataFrame
city2_del_csv = pd.read_csv('city2.csv')
```

```python
# write the data to a sqlite table
users2_del_csv.to_sql('table_users2', conn, if_exists='append', index = False)

# write the data to a sqlite table
city2_del_csv.to_sql('table_city2', conn, if_exists='append', index = False)
```

```python
## Check the data is in there in table table_users2

c.execute('''SELECT * FROM table_users2''').fetchall()
```

```python
## Check the data is in there in table table_city2

c.execute('''SELECT * FROM table_city2''').fetchall()
```
